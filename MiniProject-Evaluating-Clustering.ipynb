# Imports & Data


```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

# 1. Preprocess data
* Handle missing values (if any)
* Apply scaling (Standardization is required)


```python
df = load_wine(as_frame=True).frame
df.head()
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

display(pd.DataFrame(df_scaled, columns=df.columns).head())
```

# 2. Apply Clustering (Before PCA)
* Apply K-Means on the original data
* Record results


```python
# Initialize K-Means
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')

# Fit the model
kmeans.fit(df_scaled)

# Get the cluster assignments
clusters = kmeans.labels_

# Add clusters back to your dataframe to see the results
df['cluster_assignment'] = clusters
display(df)

print()
print("Unique clusters:", df['cluster_assignment'].unique())
```
