# Imports & Data


```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

# 1. Preprocess data
* Handle missing values (if any)
* Apply scaling (Standardization is required)


```python
# make it with out target
wine = load_wine(as_frame=True)
df = wine.data   # features only
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
# Scaling
scaler = StandardScaler()
X = scaler.fit_transform(df)

df_scaled = pd.DataFrame(X, columns=df.columns)
df_scaled.head()
```

# 2. Apply Clustering (Before PCA)
* Apply K-Means on the original data
* Record results


```python
inertia = []
for k in range(1,8):
  kmeans = KMeans(n_clusters=k,random_state=42)
  kmeans.fit(df_scaled)
  inertia.append(kmeans.inertia_)

plt.plot(range(1,8),inertia,"-o")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```

```python
# Initialize K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')

# Fit the model
kmeans.fit(X)

# Get the cluster assignments
clusters = kmeans.labels_

# Add clusters back to dataframe to see the results
df_scaled['cluster_assignment'] = clusters
display(df_scaled)

print()
print("Unique clusters:", df_scaled['cluster_assignment'].unique())
```

# 3. Apply PCA
* Reduce the data to 2 components


```python
# Apply PCA and reduce to 2 components
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
display(df_pca)

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total explained variance:", pca.explained_variance_ratio_.sum())
```

# 4. Apply Clustering (After PCA)
* Apply K-Means on the transformed data


```python
#check fo the best k
inertia = []
for k in range(1,8):
  kmeans = KMeans(n_clusters=k,random_state=42)
  kmeans.fit(df_pca)
  inertia.append(kmeans.inertia_)

plt.plot(range(1,8),inertia,"-o")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```

```python
#Apply Kmeans
kmeans = KMeans(n_clusters=3,random_state=42)
#Fit the moel
kmeans.fit(df_pca)
# Get the cluster assignments
clusters = kmeans.labels_
# Add clusters back to dataframe for resultsdisply
df_pca['cluster_assignment'] = clusters
df_pca.head()
```

# 5. Visualization
* Visualize clusters (2D using PCA components)
* Compare before vs after PCA

```python
# visualize clusters from original KMeans
plt.scatter(X_pca[:,0], X_pca[:,1],c=df_scaled["cluster_assignment"],cmap="viridis")

plt.title("KMeans on Original Data (Projected with PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
```

```python
#vizualize clusters
plt.scatter(df_pca["PC1"], df_pca["PC2"], c=df_pca["cluster_assignment"] , cmap="viridis")
plt.title("After PCA, KMeans Clustering")
plt.show()
```

# Analysis (Very Important!)

How did PCA affect clustering results?
* PCA reduced the dimensionality of the dataset from 13 features to 2 principal components while preserving most of the variance.This helped the clustering algorithm focus on the most important patterns

Did visualization become clearer?
* Visualization became clearer,because PCA reduced the data to two meaningful components, making cluster separation easier to observe.

Did you lose important information?
* Some information was lost because PCA compresses the data into fewer dimensions.However, most important variance was preserved.

Which approach is better and why?
* Clustering after PCA is often better because PCA removes noise and reduces dimensionality, which helps clustering algorithms identify patterns more effectively.

# Key Insights:


* Insight 1 : PCA reduces dimensionality while keeping most important variance
* Insight 2 : Clustering after PCA can improve visualization and cluster separation
* Insight 3 : Dimensionality reduction helps machine learning models work more efficiently
